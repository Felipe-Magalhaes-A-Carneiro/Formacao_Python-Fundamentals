# ***01 - Desafio***

Crie um programa seguindo exatamente o fluxograma a baixo.

```python
print(">>> INCIO <<<")

resposta = input("Estou com fome? ")

if resposta == "sim":
    resposta = input("Tenho comida em casa?")

    if resposta == "sim":
        print("Preparar uma refeio...")
        print("Comer a comida.")
        print(">>> FIM <<<")

    elif resposta == "no":
        print("Ir ao mercado...")
        print("Voltar para casa...")
        print("Preparar refeio...")
        print("Comer a comida.")
        print(">>> FIM <<<")

else:
    print("Ento no vou comer.")
    print(">>> FIM <<<")
```
