# ***Operadores Booleanos***

Analizam se os valores entre eles são 'True' (and) ou se ao menos um deles é 'True' (or). Enquanto isso, o operador (not) é usado para trocarmos o valor de algo que é True em False, e False em True.


----------------------------------------------------------------------------------------------------

## Seus tipos:

# ***and***

*... e ...*

Quando quisermos ANALISAR SE TODOS os VALORES que estão um dos lados dele são 'True', utilizamos and entre eles.

**Atenção:** Ele só retornará 'True' se TODOS os valores entre ele sejam 'True' (Verdadeiro).

```python
True and True # 'True' é retornado pois os valores entre ele são True
```

```python
True and False # 'False' porque afirmamos que os dois são 'True' e ele nega, o que está correto.
```

```python
False and True
```

```python
False and False
```

----------------------------------------------------------------------------------------------------

# ***or***

*... ou ...*

Quando quisermos ANALISAR SE UM DOS VALORES que está entre ele é 'True', utilizamos or entre eles.

**Atenção:** Ele só retornará 'True' se UM DOS valores entre ele sejam 'True' (Verdadeiro).

```python
True or True
```

```python
True or False
```

```python
False or True
```

```python
False or False
```

----------------------------------------------------------------------------------------------------

# ***not***

*... não é ...*

Quando quisermos AFIRMAR/MUDAR um DOS VALORES que 'not' esta acompanhando-o.

**Atenção:** Ele só retornará 'True' se UM DOS valores entre ele sejam 'True' (Verdadeiro).

```python
True
```

```python
not True
```

```python
False
```

```python
not False
```
