# ***Operadores de Comparação***

Operadores responsáveis por retornar valores como 'True' e 'False' ao compararmos dois tipos de informação.


----------------------------------------------------------------------------------------------------

## Seus tipos:

# ***==***

*igual a*

Quando quisermos saber se um dado é **IGUAL** ao o outro, utilizamos == entre eles

```python
n1 = 8
n2 = 8

n1 == n2
```

```python
n1 = 8
n2 = 10

n1 == n2
```

```python
n3 = 4
n4 = 4.0

n3 == n4 # 'True', embora sejam tipos de dados diferentes (int e float), o interpretador entende que é o MESMO VALOR.
```

```python
n5 = "4"

n4 == n5 # 'False', os dados são de tipos diferentes (int e str), o interpretador sabe que eles NÃO POSSUEM O MESMO VALOR, já que "4" é um caractere para o Python então não há valor numérico para ele
```

----------------------------------------------------------------------------------------------------

# ***!=***

*diferente de*

Quando quisermos saber se um dado é **DIFERENE** de outro, utilizamos != entre eles

```python
n1 = 8
n2 = 8

n1 != n2
```

```python
n1 = 8
n2 = 10

n1 != n2
```

----------------------------------------------------------------------------------------------------

# ***>*** 

*Maior que*

Quando quisermos saber se um dado é **MAIOR QUE** o outro, utilizamos > entre eles

```python
5 > 3
```

```python
10 > 50
```

```python
5 > 5
```

----------------------------------------------------------------------------------------------------

# ***<*** 

*Menor que*

Quando quisermos saber se um dado é **MENOR QUE** o outro, utilizamos < entre eles

```python
5 < 10
```

```python
10 < 8
```

```python
10 < 10
```

----------------------------------------------------------------------------------------------------

# ***<= e >=*** 

*Menor ou igual a* e *Maior ou igual a*

Quando quisermos saber se um dado é **MENOR OU IGUAL A**, utilizamos <=.

Quando quisermos saber se um dado é **MAIOR OU IGUAL A**, utilizamos >=.

```python
5 <= 6
```

```python
5 <= 5
```

```python
5 <= 4
```

```python
6 >= 5
```

```python
6 >= 6
```

```python
6 >= 7
```
