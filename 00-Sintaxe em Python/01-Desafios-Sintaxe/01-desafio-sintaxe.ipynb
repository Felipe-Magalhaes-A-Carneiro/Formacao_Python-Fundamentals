# 01 - Desafio
Crie um programa que:
- Pede pelo seu nome e idade;
- Dá 'Oi' para você;
- Conta quantas letras seu nome possui;
- Fala quantos anos você terá daqui a 5 anos.

```python
nome = input("Digite o seu nome: ")
idade = int(input("Digite a sua idade: "))
letras_em_nome = len(nome)
idade_mais_cinco = idade + 5

print(f"Oi, {nome}!")
print(f"Ei, {nome}, você sabia que o seu nome possui {letras_em_nome} letras?")
print(f"E que daqui a 5 anos você terá {idade_mais_cinco} anos de idade? Legal, né")
```
