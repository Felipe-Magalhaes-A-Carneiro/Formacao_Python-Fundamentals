# ***fromkeys()***

Cria um novo Objeto Dicionário, com as chaves definidas por você. Isso pode acontecer de duas formas:

- Primeira: Você quer criar as chaves, armazena-las e não colocar nenhum valor dentro dela;
- Segunda: Vocêr quer criar um CONJUNTO de chaves e um VALOR PADRÃO em cada uma delas.



```python
# Primeiro:

dicionario_modo_um = dict.fromkeys(["nome", "telefone"])
print(dicionario_modo_um)
```

```python
# Segundo:

dicionario_modo_dois = dict.fromkeys(["nome", "telefone"], "vazio")
print(dicionario_modo_dois)
```

***Obs.:*** Como este método é recomendado apenas no momento de criação de um dicionário, caso queira atualizar um dicionario com novas chaves, utilize o método update().

Caso o contrário, se tentar criar um novo Dicionário com o fromkeys() isso acontecerá:

```python
dicionario_existente = {"nome": "vazio", "telefone": "vazio"}

dicionario_existente.fromkeys(["endereco","numero"], "vazio") # Ele cria um novo dicionario com estes dados tirados do 'dicionario_existente', porém como eu não atribui em nenhuma variavel, ele executa corretamente, mas 'esquece' em seguida. Deixo claro que, a forma correta para fezer isso é utilizando o método update()
```

```python
print(dicionario_existente)
```

# ***update()***

O método update() modifica o dicionário original diretamente.

Existem duas formas de utiliza-lo:

- Opção 1: Usando update() (Recomendado para conjuntos)

```python
dicionario_existente = {"nome": "vazio", "telefone": "vazio"}

novas_chaves = dicionario_existente.fromkeys(["endereco", "numero"], "vazio")
dicionario_existente.update(novas_chaves)

print(dicionario_existente)
```

## ***Obs.:***
Segundo o Google Gemini, a forma de atribuição mais simples, seria esta:

```python
dicionario_existente["endereco"] = "vazio"
dicionario_existente["numero"] = "vazio"

print(dicionario_existente)
```

# ***get()***

Método que encontra uma chave em um dicionário. Se ela houver, retornará 'None'.

***Obs.:*** Em dicionários, não utilizamos o mesmo modo que em outras estruturas, que seria algo como 'dicionario['chave']' para encontrarmos a chave. Utilizamos o método get()


```python
contatos = {
    "felipe@gmail.com": {"nome": "Felipe","telefone": "1199998-9899"}
} 

contatos.get("endereco")
```

Podemos também dizer ao método que, se ele não encontrar a tal chave, retorne um valor para mim. Isso é feito da seguinte forma:

```python
contatos.get("endereco", "Não há nada disso aqui") 

# ou seja, dentro de argumentos, estamos defindo algo para ele encontrar, caso o contrário, retorne 'Não há nada disso aqui', mas lembrando que pode ser qualquer valor que você quiser. Essa açaao não pode ser feita quando a chave existir no dicionario
```

Perceba que ele não retorna nada, pois não há uma chave com este nome.

Mas se utilizamor quando há a chave definida no dicionário:

```python
contatos.get("felipe@gmail.com")
```

# ***setdefault()***

O método setdefault() é utilizado em dicionários para recuperar o valor de uma chave, mas com um "superpoder" adicional: se a chave não existir, ele a cria automaticamente com um valor padrão que você definir.

É uma forma mais curta e elegante de garantir que uma chave exista antes de tentar usá-la.

**Sua sintaxe:**

dicionario.setdefault(chave, valor_padrao)

1. O que acontece quando você o chama:
Se a chave já existe: O método apenas retorna o valor que já estava lá. O dicionário não é alterado.

2. Se a chave NÃO existe: O método insere a chave no dicionário com o valor_padrao fornecido e retorna esse valor.

## Comparativo: get() vs setdefault()
Muita gente confunde os dois. A diferença é fundamental:

- O get() apenas "dá uma espiada". Se a chave não existir, ele te devolve o padrão, mas não altera o dicionário original.

- O setdefault() "garante a vaga". Se a chave não existir, ele altera o dicionário inserindo a nova chave.

```python
cores_lista = ["vermelho", "azul", "vermelho", "verde"]
contador = {}

for cor in cores_lista:
    contador[cor] = contador.setdefault(cor, 0) + 1

print(contador)
```

## Por que não usar apenas contador[cor] = contador[cor] + 1?
Se você fizesse isso e a cor nunca tivesse aparecido, o Python daria um KeyError (uma Exceção!), porque você estaria tentando somar 1 a algo que ainda não existe. O setdefault evita esse erro criando a base (o zero) no momento exato da necessidade.

# ***update()***

O método update() é a ferramenta do Python para realizar atualizações em massa ou fusões de dicionários. Se o setdefault() serve para garantir uma única chave, o update() serve para injetar um conjunto inteiro de dados de uma vez.

### **1. O que ele faz na prática?**
O update() pega os pares chave-valor de outro dicionário (ou de um iterável de pares) e os "carimba" no dicionário atual.

- Se a chave já existe: O valor antigo é sobrescrito pelo novo.

- Se a chave NÃO existe: Ela é criada no dicionário.

```python
perfil = {"nome": "Lucas", "nivel": 10}
novos_dados = {"nivel": 11, "cidade": "São Paulo"}

perfil.update(novos_dados)

print(perfil)
# Saída: {'nome': 'Lucas', 'nivel': 11, 'cidade': 'São Paulo'}
```

# ***values()***

O método values() é a ferramenta que você usa quando só quer saber o que está guardado dentro das "gavetas" do seu dicionário.

## 1. O que ele retorna?

Diferente de uma lista comum, o values() retorna um objeto chamado dict_values. Ele é uma "view" (visão) dos valores.

- A "Mágica" da View: Se você alterar o dicionário original, o objeto retornado pelo values() se atualiza automaticamente! Ele não tira uma cópia estática; ele fica "olhando" para o dicionário em tempo real.

```python
estoque = {"maçãs": 10, "bananas": 5}
valores = estoque.values()

print(valores) # dict_values([10, 5])

estoque["maçãs"] = 15 # Mudamos o original
print(valores) # dict_values([15, 5]) -> A 'view' mudou sozinha!
```

## 2. Para que situações ele é recomendável?
Você deve usar o values() sempre que a chave for irrelevante para o cálculo final.

- Somatórios: Saber o total de itens no estoque sem precisar saber quais são os itens.

- Buscas de existência: Verificar se um valor específico existe em algum lugar do dicionário.

- Média/Estatística: Calcular a média de notas de alunos, por exemplo.

```python
notas = {"Alice": 9.0, "Bruno": 7.5, "Carla": 10.0}

# Quero apenas a média, os nomes não importam agora
total = sum(notas.values())
media = total / len(notas)

print(f"A média da turma é: {media}")
```

## 3. Posso usar colchetes [] no values()?

Não diretamente. Como o values() retorna uma "view" e não uma lista, você não consegue fazer valores[0].

Se você precisar acessar um valor por posição (o que é raro, já que dicionários são focados em chaves), você precisa converter para uma lista primeiro:

```python
lista_de_valores = list(notas.values())
print(lista_de_valores[0]) # Agora funciona!
```

## Por que os parênteses () estão vazios?

Note que no values() você quase nunca coloca nada dentro dos parênteses. Isso acontece porque o método não precisa de nenhuma informação extra para funcionar; o dicionário já "sabe" quais são os seus valores. Você só está dando o comando: "Ei, objeto dicionário, execute sua função de me mostrar seus valores".

# ***in***

Não é um método.

O 'in' é um Operador de Pertencimento (Membership Operator).

Diferente de um método (como .values() ou .sort()), você não o chama usando um ponto. Ele é uma palavra reservada da linguagem que fica entre dois valores, funcionando como um conector lógico.

- Sintaxe de um Método: objeto.metodo()

- Sintaxe do Operador in: item in colecao

## 2. Por que ele parece um método "disfarçado"?

Quando você escreve if "Alice" in notas:, o Python, por baixo dos panos, transforma isso em uma chamada de método:$$notas.\_\_contains\_\_("Alice")$$Ou seja: o in é apenas uma forma elegante e legível (açúcar sintático) de chamar o método mágico __contains__ que quase todo objeto iterável possui. Por isso, embora seja um operador, ele ativa um comportamento interno do objeto.

```python
notas = ["Alice"]

if "Alice" in notas:
    print("Sim")
```

Ou seja: o in é apenas uma forma elegante e legível (açúcar sintático) de chamar o método mágico __contains__ que quase todo objeto iterável possui. Por isso, embora seja um operador, ele ativa um comportamento interno do objeto.

## O erro comum no Dicionário:
Muitas pessoas escrevem if valor in dicionario.values().

- Isso funciona? Sim.

- É bom? Não. Ao usar .values(), você transforma uma busca instantânea (pela chave) em uma busca lenta (percorrendo todos os valores um por um). Se você quer saber se uma etiqueta existe, use apenas if chave in dicionario.

## 3. O in no laço for
O in também tem uma segunda vida. No for item in colecao:, ele não está apenas testando se algo existe, ele está servindo de "ponte" para o iterador pegar o próximo item.

## Resumo:
- Não é método, é um operador.

- Ele retorna sempre um Booleano (True ou False) quando usado em testes.

- Ele invoca o método mágico __contains__.
