# 1. Nível Iniciante: "O Organizador de Fila"

Objetivo: 

Crie uma função chamada gerenciar_fila(clientes) que recebe uma lista de nomes. Você deve adicionar o nome "VIP" ao início da lista e remover o último nome da lista. No final, retorne a lista alterada.

```python
clientes = ["Ana", "Bruno", "Caio"]

def gerenciar_fila(clientes):
    clientes.pop(-1)
    clientes.insert(0,"VIP")

    return clientes

gerenciar_fila(clientes)
```

# 2. Nível Fácil: "Busca e Destruição"

Objetivo: 

Crie uma função chamada limpar_lista(lista, valor) que recebe uma lista e um valor específico. A função deve verificar quantas vezes esse valor aparece. Se aparecer mais de uma vez, remova apenas a primeira ocorrência dele e retorne a lista. Se não aparecer, retorne a lista original.

```python
lista = [1, 2, 3, 2, 4]

valor = int(input("Número:"))

def limpar_lista(lista, valor):
    contagem = lista.count(valor)

    if contagem > 1:
        lista.remove(valor)
    
    return lista

limpar_lista(lista, valor)
```
