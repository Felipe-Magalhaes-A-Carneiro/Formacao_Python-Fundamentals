# ***Métodos da classe list***

Os métodos mais utilizados para a manipulação de listas

# ***append( )***

Adiciona novos valores para dentro de uma lista; inclusive até mesmo novas listas dentro de listas, dessa forma crinado o que chamamos de 'matrizes'. Além disso, por que não, anexar também uma lista dentro de outra? Com append() é possível.

***Atenção:*** Diferente do método **'extend()'**, o **append()** adiciona um novo valor por vez.

```python
lista = []

lista.append(1)
lista.append("Python")
lista.append("DIO.me")
lista.append([40, 30, 20]) # Adicona até mesmo listas
```

```python
# Anexando uma lista dentro de outra, utilizando o método append()
lista_de_numeros = [1,2,3,4,5]
print(lista_de_numeros)

lista_de_mais_numeros = [6,7,8,9,10]

lista_de_numeros.append(lista_de_mais_numeros)
print(lista_de_numeros)
```

```python
print(lista)

for i in lista:
    print(i)
```

# ***Obs.:***

É possível modificar dados dentro de uma lista, sem o uso do ***'append()'***. Embora, segundo os meus estudos o recomendado mesmo é o uso desse método, mas a seguir vou mostrar como isso é feito; muito semelhante a forma que fazemos um 'acesso direto' na lista:

```python
lista[0] = 10

print(lista)
```

# ***clean( )***

Usado para limpar tudo o que há em uma lista

```python
lista.clear()

print(lista)
```

# ***copy( )***

A lista é um objeto mutavél, ou seja, ela pode ser alterada. Com esse método, é possível criar uma nova lista, tendo uma nova instância (portanto não é a mesma lista que foi criada).

```python
# Trazendo novos valores a lista que foi limpa
lista.append(2)
lista.append("Felipe")
lista.append("Magalhães")
lista.append([31, 32, 33]) # Adicona até mesmo listas

print(lista)
```

```python
lista2 = lista.copy()

# Comprovação que as listas são diferentes, embora uma é uma cópia, de todos os dados da outra
print(id(lista), lista)
print(id(lista2), lista2)
```

# ***count()***

Método responsável por fazer a contagem de quantas vezes um valor repetido aparece em uma lista.

***Aplicação:*** este método necessita de um argument e, nele coloque o valor que deseja que ele encontre e conte quantas vezes este valor está na lista.

```python
cores = ["Vermelho", "Azul", "Amarelo", "Azul", "Vermelho", "Azul"]


print("Vermelho:", cores.count("Vermelho"))
print("Azul:", cores.count("Azul"))
print("Amarelo:", cores.count("Amarelo"))
```

# ***extend()***

Método muito utilizado. Neste é possível adicionar vários valores dentro de uma lista, inclusive pode adicionar os valores de uma lista em uma outra lista. 

***Aplicação:*** Este método necessita que dentro do seu argumento também contenham colchetes [] e dentro deste os valores que você quer.

***Obs.:*** 
- Diferente de append() que é um por vez.
- Ele não elimina valores duplicados

```python
linguagens = ["Python", "JavaScript", "C"]

print(linguagens)

linguagens.extend(["Java", "Rubi"])
print(linguagens)
```

```python
# adicionando valores de uma lista em outra lista:
linguagens_baixo_nivel = [ "C++" , "Rust" ]

linguagens.extend(linguagens_baixo_nivel)

print(linguagens)
```

# ***index()***

Responsável por trazer a posição de cada valor por meio de seu index, começando do 0.

***Obs.:*** No caso de valores duplicados, ele apenas irá trazer o primeiro valor que está na sequência da lista, ignorando os duplicados que virão em sequência.

```python
linguagens.index("Python")
```

```python
linguagens.index("Java")
```

# ***pop()***

Uma de duas formas para remover um dado de uma lista. Ao usa-lo sem uma argumentação, ele remove o último dado em uma lista. Uma vez definido dentro de seu argumento um index, ele excluirá este index definido da lista

```python
linguagens.pop()

print(linguagens)
```

```python
linguagens.pop(1)

print(linguagens)
```

# ***remove()***

Segunda de duas formas para remover um dado de uma lista. Em caso de dados duplicados, ele removerá a primeira ocorrência (index) duplicada que houver na lista

```python
# Adicionar um valor duplicado:
linguagens.append("C")

print(linguagens)
```

```python
linguagens.remove("C")

print(linguagens)
```

# ***reverse()***

Retorna a lista, mas invertida. 

```python
linguagens.reverse()

print(linguagens)
```

# ***sort()***

Retorna a lista em ordem alfabética. 

***Aplicação:***
- Se não houver argumento, ele ordenará por ordem alfabética;
- Se houver dentro do argumento 'reverse=True', será por ordem alfabética, mas será o contrário;
- Se houver dentro do argumento 'key=lambda x: len(x)', será por ordem do tamanho sendo crescente;
- Se houver dentro do argumento 'key=lambda x: len(x), reverse=True', será por ordem do tamanho; sendo decrescente.

```python
linguagens.sort()
print(linguagens)
```

```python
linguagens.sort(reverse=True)
print(linguagens)
```

```python
linguagens.sort(key=lambda x: len(x))
print(linguagens)
```

```python
linguagens.sort(key=lambda x: len(x), reverse=True)
print(linguagens)
```

# ***Sorted()***

Função que faz o mesmo que o método sort(), mas a sua forma de usar é um pouco diferente:

```python
print(linguagens)

sorted(linguagens)
```

```python
print(linguagens)

sorted(linguagens, key=lambda x: len(x))
```

```python
print(linguagens)

sorted(linguagens, key=lambda x: len(x), reverse=True)
```

# ***len()***

Função que retorna o tamanho de uma lista. 

***Obs.:*** Aqui é feita utilizando o número de quantos index a lista selecionada possuí.

```python
print(linguagens)
len(linguagens)
```
