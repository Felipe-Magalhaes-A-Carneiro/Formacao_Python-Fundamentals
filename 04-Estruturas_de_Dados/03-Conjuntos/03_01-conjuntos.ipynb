# ***Conjuntos em Python***

Antes de tudo é necessário entender o funcionamento da estrutura de dados set.

## ***Criando Sets (conjuntos, coleção)***

Um set é uma função e uma coleção que não possui objetos repetidos, ou seja, apenas elementos únicos. Usamos sets para representar conjuntos matemáticos ou eliminar itens duplicados de um interável.

**Aplicação:**
Set é um construtor, portanto ele é uma função que necessita de argumentos.

## ***1.1 - Obs.:***
Set é utilizado para remover valores duplicados também. Podemos utilizado, por exemplo, na criação de uma lista.

Perceba que no exemplo abaixo, a função tem em seu campo de argumento a criação da lista que, esta por sua vez possuí valores repetidos, ao executarmos o set elimina os valores repetidos, mantêm apenas um destes e o organiza. 

Detalhe: **No caso de Strings**, ele prioriza o primeiro valor que não é repetido e então, traz os demais de forma **ordenada e alfabética**

```python
set(["Vermelho", "Azul", "Amarelo", "Azul", "Vermelho", "Azul", "Marrom", "Cinza"])
```

```python
set(["Vermelho", "Azul", "Azul", "Vermelho","Amarelo", "Azul"])
```

## ***1.2 - Obs.:***
**No caso dos inteiros**, ele remove os repetidos, mantêm apenas um deles e então, ordena os valores de forma crescente

```python
set([0, 1, 2, 1, 4, 5, 5, 7, 8, 9, 10, 10, 10])
```

```python
set([10, 1, 2, 1, 4, 5, 5, 7, 8, 9, 10, 10])
```

```python
numeros = set([10, 1, 2, 1, 4, 5, 5, 7, 8, 9, 10, 10])

print(numeros)
```

```python
set("abacaxi")
```

```python
linguagens = set(("Python", "JavaScript", "C","JavaScript", "Java")) # set e uma tupla. Perceba que ele traz os valores, mas sua ordenação é alfabética de trás pra frente 

print(linguagens)
```

# ***Acessando os dados***

Não ocorre utilizando o 'acesso direto' como em listas e em tuplas; Conjuntos em Python não suportam indexação e nem fatiamento, caso queira acessar os seus valores **é necessário converter o set (conjunto) para uma lista**.

```python
linguagens_lista = list(linguagens) # Conversão para lista

print(linguagens_lista)
```

```python
linguagens_lista[2]
```

# ***Interar sets***

A forma mais comum para percorrer os dados de um set(conjuntos) é utilizando também o ***for***.

```python
for numero in numeros:
    print(numero)
```

```python
for i, numero in enumerate(numeros):
    print(f"Posição do index: {i} é {numero}")
```
