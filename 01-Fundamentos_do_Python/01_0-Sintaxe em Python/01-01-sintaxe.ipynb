# 1. Por que o ponto (.) antes do método?

O ponto é chamado de operador de acesso.
 
Em Python, quase tudo é um objeto (uma estrutura que guarda dados e ações). Imagine que um objeto é uma "caixa" com ferramentas dentro.

O objeto vem antes do ponto.

A ação (método) vem depois do ponto.

Exemplo: Se você tem uma frase (string), ela é um objeto do tipo texto. O método upper() é uma ferramenta que pertence a esse texto.

```python
nome = "python"

print(nome.upper()) # "Acesse o objeto 'nome' e use a ferramenta 'upper' dele"
```

# 2. Por que os parênteses ()?

Os parênteses são o sinal de execução. Eles dizem ao Python: "Não apenas olhe para esta função, execute-a agora".

- Sem parênteses: Você está apenas referenciando o nome da função (como se estivesse lendo o manual da ferramenta).

- Com parênteses: Você está ligando a ferramenta.

O que vai dentro deles?
Dentro dos parênteses vão os **argumentos** (ou parâmetros). São as informações extras que a função precisa para trabalhar.

- print() precisa saber o que imprimir.

- round(10.55, 1) precisa saber o número e quantas casas decimais você quer

# 3. Colchetes [] dentro de Parênteses ()?

Quando você vê algo como funcao([1, 2, 3]), isso acontece porque você está passando uma Lista como argumento para a função.

- () Parênteses: Comando de execução da função.

- [] Colchetes: Definem uma Lista (uma coleção de dados ordenada).

Por que usar assim? Imagine a função sum(), que soma números. Ela não soma números "soltos" bem, ela prefere receber um pacote (lista) de números:

```python
# Errado: sum(1, 2, 3) -> O Python vai achar que o 2 e o 3 são outras configurações

# Correto:
sum([1, 2, 3]) # Você entrega um "pacote" (lista) fechado para a função
```

# 4. Boas Práticas e Métodos em Python

Para escrever um código "Pythonico" (elegante e eficiente), siga estes padrões:

## Funções vs. Métodos
- Função: É independente. Ex: len("texto").

- Método: É preso a um objeto. Ex: "texto".upper().

## Regras de Ouro (PEP 8)
1. Nomes de funções: Use sempre letras minúsculas e sublinhados (snake_case). Ex: calcular_media_final().

2. Identação: Python usa espaços (4 espaços) para saber o que está dentro de uma função. Se não indentar, o código quebra.

3. Docstrings: Sempre explique o que sua função faz logo na primeira linha dela.

```python
def saudar_usuario(nome):
    """Exibe uma saudação personalizada."""
    return f"Olá, {nome}!"

saudar_usuario("Felipe")
```
